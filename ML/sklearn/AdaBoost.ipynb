```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
X,y=make_classification(n_samples=1000,n_features=50,n_informative=30,n_clusters_per_class=3)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
clf=DecisionTreeClassifier(random_state=11)
clf.fit(X_train,y_train)
accuracy=clf.score(X_test,y_test)
print("Decision tree accurracy: %s" %accuracy)
```

```python
#When an argument for the base_estimator parameter is not passed, the default DecisionTreeClassifier is Uesd
clf=AdaBoostClassifier(n_estimators=50,random_state=11) 
clf.fit(X_train,y_train)
# accuracy.append(clf.score(X_test,y_test))
```

```python
plt.title('Ensemble Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Number of base estimators in ensemble')
plt.plot(range(1,51),[accuracy for accuracy in clf.staged_score(X_test,y_test)])
```
