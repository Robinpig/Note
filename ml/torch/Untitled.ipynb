```python
import requests
```

```python
target="https://baidu.com/"
req=requests.get(url=target)
```

```python
print(req.text)
```

```python
biqu="https://www.biqugexx.com/119_119677/33450994.html"
req=requests.get(url=biqu)
print(req.text)
```

```python
from bs4 import BeautifulSoup
html = req.text
bf = BeautifulSoup(html)
texts = bf.find_all('div',id='content' )
print(texts[0].text.replace('<br/><br/>','\n'))
```

```python
server="https://www.biqugexx.com"
list="https://www.biqugexx.com/119_119677/"
req = requests.get(url = list) 
html = req.text
div_bf = BeautifulSoup(html)
div = div_bf.find_all('div', id='list')
a_bf = BeautifulSoup(str(div[0]))
a = a_bf.find_all('a')
for each in a:
    print(each.string, server + each.get('href'))
```
